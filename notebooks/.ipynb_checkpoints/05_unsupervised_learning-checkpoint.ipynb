```python
# 05 - Aprendizaje no supervisado (Clustering)

En este notebook se analiza el desempeño de distintos algoritmos de aprendizaje no supervisado aplicados al dataset procesado en el proyecto. El objetivo es:

- Agrupar observaciones similares mediante *clustering*.
- Comparar el rendimiento de varios algoritmos (K-Means, DBSCAN y GMM).
- Analizar el perfil de cada cluster en función de las variables numéricas.
- Visualizar la estructura de los clusters en un espacio reducido (PCA 2D).
```

```python
from pathlib import Path
import yaml


cwd = Path(".").resolve()

if (cwd / "conf" / "base" / "catalog.yml").exists():
    project_path = cwd
elif (cwd / "spaceflights" / "conf" / "base" / "catalog.yml").exists():
    project_path = cwd / "spaceflights"
else:

    project_path = Path(r"C:\ML\spaceflights")

catalog_path = project_path / "conf" / "base" / "catalog.yml"

print("Project path:", project_path)
print("Catalog path:", catalog_path)

with open(catalog_path, "r", encoding="utf-8") as f:
    catalog_cfg = yaml.safe_load(f)

{k: v for k, v in catalog_cfg.items() if "unsupervised" in k}
```

```python
import pandas as pd


labels_fp = catalog_cfg["unsupervised_cluster_labels"]["filepath"]
metrics_fp = catalog_cfg["unsupervised_cluster_metrics"]["filepath"]
clusters_fp = catalog_cfg["unsupervised_model_input_with_clusters"]["filepath"]

labels_path = project_path / labels_fp
metrics_path = project_path / metrics_fp
df_clusters_path = project_path / clusters_fp

print(labels_path)
print(metrics_path)
print(df_clusters_path)

labels = pd.read_parquet(labels_path)
metrics = pd.read_csv(metrics_path)
df_clusters = pd.read_parquet(df_clusters_path)

type(labels), type(metrics), type(df_clusters)

```

```python
## Vista rápida de los datos cargados
print("Métricas de clustering:")
display(metrics.head())

print("\nDataset enriquecido con cluster_label:")
display(df_clusters.head())

print("\nShape del dataset enriquecido:", df_clusters.shape)
print("Clusters disponibles:", df_clusters["cluster_label"].unique())
```

```python
## Métricas de evaluación por algoritmo

A continuación se comparan los algoritmos de clustering utilizando métricas estándar:

- Silhouette 
- Calinski–Harabasz 
- Davies–Bouldin

Estas métricas permiten evaluar qué tan compactos y separados quedan los clusters.
```

```python
metrics_sorted = metrics.sort_values("silhouette", ascending=False)
metrics_sorted
```

```python
En la tabla anterior se observa que:
Cada métrica favorece a un algoritmo distinto, pero la más importante para clustering general es Silhouette.
El algoritmo que mejor desempeño global presenta es DBSCAN, porque:
Tiene la mayor Silhouette.
Forma agrupaciones más naturales según densidad.
Maneja bien outliers sin obligar a formar clusters esféricos.
```

```python
##Perfil de los clusters
cluster_col = "cluster_label"

feature_cols = [
    col
    for col in df_clusters.columns
    if col != cluster_col and pd.api.types.is_numeric_dtype(df_clusters[col])
]

len(feature_cols), feature_cols[:10]

```

```python
cluster_profile_describe = df_clusters.groupby(cluster_col)[feature_cols].describe().round(3)
cluster_profile_describe
```

```python
Los clusters representan tipologías de objetos cercanos a la Tierra, separándose principalmente por:
Tamaño/ diámetro
Velocidad relativa
Distancia mínima a la Tierra (miss_distance)
Año de descubrimiento y magnitud absoluta
```

```python
Cluster 0
-Es el grupo más numeroso (10.5K registros).
-Valores medios altos en velocidad relativa y distancia mínima.
-Objetos con mayor tamaño promedio.
-Representa los objetos "normales" del dataset.

Cluster 1
-Segundo más numeroso (4.6K).
-Menores velocidades y distancias.
-Menor diámetro.
-Representa objetos más pequeños y más cercanos.

Cluster 2
-3.4K registros.
-Velocidades más altas.
-Diámetros grandes.
-Representa objetos potencialmente más peligrosos, rápidos y grandes.
    
Cluster 3
-Muy pequeño (~300 registros).
-Mucha variabilidad y valores extremos.
-Coincide con “outliers” o casos excepcionales.
-DBSCAN suele asignar estos puntos como ruido o grupos muy diferentes del resto.
```

```python
##Visualización de clusters en espacio reducido (PCA 2D)
Para complementar el análisis numérico, se aplica Análisis de Componentes Principales (PCA) y se proyectan las observaciones.

```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

X = df_clusters[feature_cols].fillna(0)

pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X)

df_clusters["PC1"] = X_pca[:, 0]
df_clusters["PC2"] = X_pca[:, 1]

print("Varianza explicada por PC1 y PC2:", pca.explained_variance_ratio_)

plt.figure(figsize=(8, 6))
for label, group in df_clusters.groupby(cluster_col):
    plt.scatter(group["PC1"], group["PC2"], alpha=0.6, label=f"Cluster {label}")

plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("Clusters en espacio PCA 2D")
plt.legend()
plt.show()
```

```python
-PC1 = 54.4% de la varianza
-PC2 = 45.5% de la varianza
Por lo tanto, el plano 2D conserva 100% de la información :D!!

Los clusters 0, 1 y 2 se traslapan bastante:
Esto confirma que sus diferencias son numéricas pero no necesariamente geométricas en 2D.

El cluster 3 aparece más separado
Esto refuerza la idea de que contiene valores atípicos o poco frecuentes.

La distribución sugiere que muchos de los features están correlacionados.
```
